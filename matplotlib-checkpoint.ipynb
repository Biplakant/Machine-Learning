# Python Matplotlib Histogram

```python
#Matplotlib histogram is a representation of numeric data in the form of a rectangle bar. Each bar shows some data, 
```

```python
#Syntax: plt.hist(x,bins=None,range=None,density=None,weights=None,cumulative=False,bottom=None,
#histtype=’bar’,align=’mid’,orientation=’vertical’,rwidth=None,log=False,color=None,label=None,stacked=False,normed=None,*,
#data=None,**kwargs,)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(15,45,100) #18= start from and 45 end and 100 total number of values
py_student_age=np.random.randint(15,40,100)
print(ml_student_age)
print(py_student_age)
```

```python

plt.hist(ml_student_age,rwidth=0.8,histtype="bar",orientation="horizontal",color="b",label="ml_studet")

plt.title("ml_student_age")
plt.xlabel("student age catogory")
plt.ylabel("number of student")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]# category of ML students age on x axis
plt.figure(figsize=(20,6))## size of histogram in 20:6 format
 
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.8,histtype="bar",orientation="horizontal",color=["b","y"],label=["ML studet","py student"])

plt.title("ml_student_age")
plt.xlabel("student age catogory")
plt.ylabel("number of student")
plt.legend()
plt.show()
```

```python
#Matplotlib Bar Chart
#To visualize value associated with categorical data in the bar format use matplotlib bar chart plt.bar() or plt.barh() methods.
```

```python
from matplotlib  import  style #background style
```

```python
class_1=["python","R","AI","ML","DS"]
class_2=["30","10","20","25","10"]
class_3=[35,5,40,14,16]
```

```python
#To plot bar chart using the matplotlib python library, use plt.bar() or plt.barh() methods.

#Syntax: plt.bar(x,height,width=0.8,bottom=None,*,align=’center’,data=None,**kwargs,)
```

```python
plt.bar(class_1,class_2)## Plot vertical Bar Chart
```

```python
plt.barh(class_1,class_2)## Plot horizontal bar chart 
```

```python
plt.bar(class_1,class_2,width=0.2,align="edge",color="y",edgecolor="m",linewidth=5,alpha=0.8,linestyle="--",label="class_1")
style.use("ggplot")
plt.xlabel("class")
plt.ylabel("class_1")
plt.legend(class_1)
plt.title("classs vs class_1")
```

```python
plt.figure(figsize=(15,5))
plt.bar(class_1,class_2,align="edge",color="g",edgecolor="m",linewidth=5,alpha=0.6,linestyle=("--"),label="class_meet")
plt.bar(class_1,class_3,align="edge",color="m",edgecolor="k",linewidth=1,alpha=0.3,linestyle=("--"),label="python_meet")
plt.xlabel("class",fontsize=25)
plt.ylabel("class1",fontsize=25)
plt.legend()
plt.title("class vs class1",fontsize=25)
plt.xticks(class_1,class_2,rotation=20)
plt.show()
```

```python
#Matplotlib Pie Chart
#The matplotkib plt.pie() function help to plot pie chart of given numeric data with labels.
#Syntax: plt.pie([‘x’,‘explode=None’,‘labels=None’,‘colors=None’,‘autopct=None’,‘pctdistance=0.6’,‘shadow=False’,
#‘labeldistance=1.1’,‘startangle=None’,‘radius=None’,‘counterclock=True’,‘wedgeprops=None’,‘textprops=None’,‘center=(0, 0)’
#‘frame=False’,‘rotatelabels=False’,‘*’,‘data=None’],)
```

```python
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 
           'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
explode=[0,1,0,0,0] #To slice the perticuler section(ek particular slice me alag karne ke liye)
colors=["c","b","r","y","g"]# Color of each section
#autopct means auto percentage.perticualr slice ka % kitna hai ..0.2=last 2 digit,f=flot
# Radius to increase or decrease the size of pie chart 
# Start angle of first section
textprops={"fontsize":15}# Font size of text in pie chart
plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,startangle=0,
        textprops =textprops)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
explode=[0,1,0,0,0] 
colors=["c","b","r","y","g"]
textprops={"fontsize":15}
#pctdistance=0.7 percentage ka distance
#labeldistance = 1.6 lables distance

plt.pie(class1_students,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,startangle=0,
        textprops =textprops,pctdistance=0.7,labeldistance = 1.6)
plt.legend()
plt.show()
```

```python
import pandas as pd
playstore=pd.read_csv("googleplaystore.csv")
playstore.head()
```

```python
playstore.columns
```

```python
x=playstore["Rating"]
y=playstore["Reviews"]
print(x)
print(y)
```

```python
#Syntax : plt.scatter(x,y,s=None,c=None,marker=None, cmap=None,norm=None,vmin=None,vmax=None,alpha=None,linewidths=None,
#verts=None, marker edgecolors=None,*,data=None,**kwargs,)
plt.figure(figsize=(15,6))
plt.scatter(x,y,color="r",marker="+",s=5,alpha=0.5, linewidths=10,
           edgecolors="g" )
plt.scatter(x,playstore["Installs"], c = "y", marker = "o", s = 100, alpha=0.5, linewidths=10,
           edgecolors="c" )
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.title("google play store")
plt.show()
plt.savefig("scater")
```
