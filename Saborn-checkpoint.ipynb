```python
#Load DataFrame from GitHub Seaborn Repository
tips_df=sns.load_dataset("tips")
tips_df.shape
tips_df.head()
```

```python
#Draw line plot of total_bill and tip
sns.lineplot("total_bill","tip",data=tips_df)
```

```python
sns.lineplot("size","total_bill",data=tips_df)
```

```python
#hue => Get separate line plots for the *third categorical* variable.#
#In the above graph draw relationship between size (x-axis) and total-bill (y-axis)
#.Now, plotting separate line plots for Female and Male category of variable sex.
#style => Give style to line plot, like dashes. Different for each line plot
#palette => Give colormap for graph.
plt.figure(figsize=(15,6))#figure size with ratio 
sns.set(style="darkgrid")# background darkgrid style of graph 

sns.lineplot("size","total_bill",data=tips_df,hue="sex",style="sex",palette="hot",dashes=False,markers=["o",">"],
             legend="full",)

plt.xlabel("Size",fontsize=20)
plt.ylabel("total_bill",fontsize=20)
plt.title("Line plot",fontsize=20)
plt.show()
```

```python
#kind : {"hist", "kde", "ecdf"}Approach for visualizing the data
#bins=how many parts to divide
#height : scalar.Height (in inches) of each facet
#aspect : scalar.Aspect ratio of each facet
plt.figure(figsize=(20,6))
sns.displot(tips_df["total_bill"],bins=10,kde=True,rug=True,color="y",height=5,aspect=1,label="Total")
plt.xlabel("total") #seaborn perform automatic xlabel so we can used xlable 
plt.legend()
plt.show()
```

```python
titanic_df=sns.load_dataset("titanic")
titanic_df.head()
```

```python
plt.figure(figsize=(16,5))
sns.scatterplot(x="age",y="fare",data=titanic_df)
```

```python
#hue: Pass value as a name of variables or vector from DataFrame, optional
#palette: Pass value as a palette name, list, or dict, optional
"""To change the color of the seaborn scatterplot. While using the palette, first mention hue parameter. """
plt.figure(figsize=(16,5))
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",hue_order=["female","male"],palette="pink_r")
plt.show()
```

```python
#size: Pass value as a name of variables or vector from DataFrame, optional
#sizes: Pass value as a list, dict, or tuple, optional
#size_order: Pass value as a list, optional.Same like hue_order size_order parameter change the size order of size variable
plt.figure(figsize=(16,5))
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",size="who",sizes=(20,50),size_order=["child","woman","man"])
plt.show()
```

```python
#style: Pass value as a name of variables or vector from DataFrame, optional
#style_order: Pass value as a list, optional
plt.figure(figsize=(16,5))
sns.scatterplot(x="age",y="fare",data=titanic_df,style="who",style_order=["child","woman","man"])
plt.show()
```

```python
#markers: Pass value as a boolean, list, or dictionary, optional
#egend: Pass value as “full”, “brief” or False, optional
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who", markers= ['3','1',3],legend=False)
```

```python
# Seaborn Scatter Plot Example 1 created by www.IndianAIProduction.com
 
# Import libraries
import seaborn as sns # for Data visualization
import matplotlib.pyplot as plt # for Data visualization
 
sns.set() #  set background 'darkgrid' 
 
#Import 'titanic' dataset from GitHub Seborn Repository
titanic_df = sns.load_dataset("titanic")
 
plt.figure(figsize = (16,9)) # figure size in 16:9 ratio
 
# create scatter plot
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex", palette = "magma",size = "who",
                sizes = (50, 300))
 
plt.title("Scatter Plot of Age and Fare", fontsize = 25) # title of scatter plot
plt.xlabel("Age", fontsize = 20) # x-axis label
plt.ylabel("Fare", fontsize = 20) # y-axis label
plt.savefig("Scatter Plot of Age and Fare") # save generated scatter plot at program location
plt.show() # show scatter plot
```

```python
# Import libraries
import seaborn as sns # for Data visualization
import matplotlib.pyplot as plt # for Data visualization
 
sns.set() #  set background 'darkgrid' 
 
#Import 'titanic' dataset from GitHub Seborn Repository
titanic_df = sns.load_dataset("titanic")
 
plt.figure(figsize = (16,9)) # figure size in 16:9 ratio
 
# create scatter plot
sns.scatterplot(x = "who", y = "fare", data = titanic_df, hue = "alive", style = "alive", palette = "viridis",size = "who",
                sizes = (200, 500))
 
plt.title("Scatter Plot of Age and Fare", fontsize = 25) # title of scatter plot
plt.xlabel("Age", fontsize = 20) # x-axis label
plt.ylabel("Fare", fontsize = 20) # y-axis label
plt.savefig("Scatter Plot of Age and Fare") # save generated scatter plot at program location
plt.show() # show scatter plot
```
