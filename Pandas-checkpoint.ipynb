```python
df1=pd.DataFrame({"ID":[1,2,3,4],
                  "class":[9,10,11,12]})
dh2=pd.DataFrame({"ID":[1,2,3,4],
                  "Name":["alpha","beta","gamma","pamma"]})
print(df1)
print(dh2)
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],
                  "class":[9,10,11,12]})
dh2=pd.DataFrame({"ID":[1,2,3,55],
                  "Name":["alpha","beta","gamma","pamma"]})
print(df1)
print(dh2)
```

```python
pd.merge(dh2,df1) # id me index number 3 me same ni tha to pandas automatic remove kar diya
```

```python
pd.merge(df1,dh2,how="left") # output mujhe left data frame se hisab se mila hai. 
#inner mera default rahata hau 
```

```python
pd.merge(df1,dh2,how="right")
```

```python
pd.merge(df1,dh2,how="outer") #provide all values
```

```python
pd.merge(df1,dh2,how="outer",indicator=True)#each row ka information janne ke liye malum karte hau
```

```python
#Pandas Join() Method in Python
df1=pd.DataFrame({"A":[1,2,3],
                 "B":[10,20,30]})
df2=pd.DataFrame({"C":[4,5,6],
                 "D":[40,50,60]})
print(df1)
print(df2)
```

```python
df1.join(df2)
```
