```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import  OneHotEncoder,OrdinalEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
## adding simple imputer to Ca   and Thal col
si=SimpleImputer(strategy="most_frequent")
Ca1=si.fit_transform(df[['Ca']])
```

```python
ca2=pd.DataFrame(Ca1,columns=['Ca'])
```

```python
si=SimpleImputer(strategy="most_frequent")
thal_new=si.fit_transform(df[['Thal']])
```

```python
pd.DataFrame(thal_new,columns=['Thal'])
```

```python
# Ordinalencoding -> ChestPain and Thal
oe=OrdinalEncoder(categories=[['asymptomatic','nonanginal','nontypical','typical']])
CheasPain_new=oe.fit_transform(df[['ChestPain']])
CheasPain_new
```

```python
CheasPain2=pd.DataFrame(CheasPain_new,columns=['cheast'])
```

```python
CheasPain2
```

```python
oe=OrdinalEncoder(categories=[['normal','reversable','fixed']])
thal2=oe.fit_transform(thal_new,[['Thal']])
thal2
```

```python
thal1=pd.DataFrame(thal2,columns=['thal'])
thal1
```

```python
#lable encoding
lr=LabelEncoder()
AHD=lr.fit_transform(df['AHD'])
```

```python
AHD_new=pd.DataFrame(AHD,columns=['AHD'])
AHD_new
```

```python
X=df.drop(['ChestPain','Thal','AHD','Ca'],axis=1)
```

```python
X
```

```python
df_new=pd.concat([X,AHD_new,thal1,CheasPain2,ca2],axis=1)
```

```python
df2=df_new.copy()
```

```python
X=df2.drop('AHD',axis=1)
y=df['AHD']
print('shape of X=',X.shape)
print('shape of y=',y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2)
print('shape of X_train=',X_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_train=',y_train.shape)
print('shape of y_test=',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
```

```python
y_pre1=clf1.predict(X_test)
y_pre2=clf2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy of Logistic Regression",accuracy_score(y_test,y_pre1))
print("Accuracy of Decision Trees",accuracy_score(y_test,y_pre2))
```

```python
confusion_matrix(y_test,y_pre1)
```

```python
print("Logistic Regression Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pre1),columns=list(range(0,2)))
```

```python
print("Decision Tree Confusion Matrix\n")
pd.DataFrame(confusion_matrix(y_test,y_pre2),columns=list(range(0,2)))
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_test
result['Logistic Regression Prediction'] = y_pre1
result['Decision Tree Prediction'] = y_pre1
```

```python
result.sample(10)
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
print("For Logistic regression Model")
print("-"*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pre1),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("Recall - ",recall_score(y_test,y_pre1,average='binary',pos_label='Yes'))
print("F1 score - ",f1_score(y_test,y_pre1,pos_label='Yes'))
print("Precision - ",precision_score(y_test,y_pre1,pos_label='Yes'))
```

```python
print("For DT Model")
print("-"*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pre2),columns=list(range(0,2)))
print(cdf)
print("-"*50)
print("Precision - ",precision_score(y_test,y_pre2,pos_label='Yes'))
print("Recall - ",recall_score(y_test,y_pre2,pos_label='Yes'))
print("F1 score - ",f1_score(y_test,y_pre2,pos_label='Yes'))
```
