```python
plt.figure(figsize=(20,10))
sns.heatmap(train_dataset.isnull())
```

```python
train_null_var=train_dataset.isnull().sum()/train_dataset.shape[0]*100  #find all null values in percentage.

train_null_var
```

```python
drop_columns  = train_null_var[train_null_var >17].keys()
drop_columns
```

```python
df2_drop_clm=train_dataset.drop(columns=drop_columns)
df2_drop_clm.shape
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2_drop_clm.isnull())
```

```python
drop_clm_rows=df2_drop_clm.dropna()
drop_clm_rows.shape
```

```python
sns.heatmap(drop_clm_rows.isnull())
```
